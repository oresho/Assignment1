```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
file = pd.read_csv("/Users/oreoluwasomuyiwa/Downloads/data.csv")
Africa = np.array(["Algeria", "Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cameroon", "Cape Verde", "Central African Republic", "Chad", "Comoros", "Congo", "Congo, Democratic Republic of", "Djibouti", "Egypt", "Equatorial Guinea", "Eritrea", "Ethiopia", "Gabon", "Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Ivory Coast", "Kenya", "Lesotho", "Liberia", "Libya", "Madagascar", "Malawi", "Mali", "Mauritania", "Mauritius", "Morocco", "Mozambique", "Namibia", "Niger", "Nigeria", "Rwanda", "Sao Tome and Principe", "Senegal", "Seychelles", "Sierra Leone", "Somalia", "South Africa", "South Sudan", "Sudan", "Swaziland", "Tanzania", "Togo", "Tunisia", "Uganda", "Zambia", "Zimbabwe"])
data = file[file['Country or Area'].isin(Africa)]
data.set_index('Country or Area', inplace=True)


print(data)
```

```python
for year in range(2004, 2015):
    data[data["Year"]== year].plot.barh(y='Value', figsize=(12.5, 12.5), title=year, fontsize=12.5)
```

```python
from statistics import mean, median, mode, stdev

letsgo = []
letsgo2 = []
letsgo3 = []
letsgo4 = []
letsgo5 = []
for year in range(2004,2015):
    Year_values = pd.DataFrame(data[data["Year"] == year].Value)
    newpd1 = Year_values.mean()
    newpd2 = Year_values.median()
    #newpd3 = Year_values.mode()
    newpd4 = Year_values.var()
    newpd5 = Year_values.std()

    letsgo.append(newpd1)
    letsgo2.append(newpd2)
    #letsgo3.append(newpd3)
    letsgo4.append(newpd4)
    letsgo5.append(newpd5)

print(newpd4)
```

```python
plt.plot(letsgo, label = 'Mean', linestyle = 'dotted')
plt.plot(letsgo2, color = 'r', label = 'Median', linestyle = 'dashed')
plt.plot(letsgo5, color = 'y', label = 'Standard Deviation')

plt.title('Combined Statistics of All African Countries against Year')

plt.xlabel('Years')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.plot(letsgo4, label = 'Variance')
plt.xlabel('Years')
plt.ylabel('Value')
plt.title('Combined Variance of All African Countries against Year')

plt.legend()
plt.show()
```
